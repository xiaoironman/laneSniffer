```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sample.jpg')
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY))
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
edges = cv2.Canny(gray,250,500) 
```

```python
plt.imshow(edges)
```
